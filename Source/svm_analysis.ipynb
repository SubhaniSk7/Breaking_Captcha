```python
import glob
from sklearn.svm import SVC
from sklearn.externals import joblib
import numpy as np
from skimage import io

from sklearn.metrics import accuracy_score,f1_score,classification_report
from sklearn.metrics import confusion_matrix
```

```python
My_SVM = joblib.load('SVM.sav')
```

```python
def normalise(l):
    ret = []
    c = len(l)
    for i in range(c):
        for j in range(c):
            ret.append(l[i][j])
    return np.array(ret)
```

```python
def get_predicted_value_and_score_for_single_image(img_folder_location):
    label = img_folder_location.split('/')[-1][0]
    image = io.imread(img_folder_location + "/Char0.jpg")
    normalised_img = normalise(image)
    ls_i = np.array([normalised_img])
    v = My_SVM.predict(ls_i)
    print("ground truth : " + label + "   predicted value : " + chr(v+65))
    if(label != chr(v+65)):
        print('-----------------------')
    return label, chr(v+65)
```

```python
def get_scores_for_all_test_images():
    file_paths = []
    y_actual=[]
    y_predicted=[]
    for path in glob.glob("./Test1char/*"):
        ground_truth,predicted_value = get_predicted_value_and_score_for_single_image(path)
        y_actual.append(ground_truth)
        y_predicted.append(predicted_value)
    
    print(accuracy_score(y_actual,y_predicted))
    print('f1_score:',f1_score(y_actual,y_predicted,average='macro'))
    print('f1_score:',f1_score(y_actual,y_predicted,average='micro'))
    print('f1_score:',f1_score(y_actual,y_predicted,average='weighted'))
    print('f1_score:',f1_score(y_actual,y_predicted,average=None))
    print(classification_report(y_actual, y_predicted))
    print(confusion_matrix(y_actual, y_predicted))
    
get_scores_for_all_test_images()
```
